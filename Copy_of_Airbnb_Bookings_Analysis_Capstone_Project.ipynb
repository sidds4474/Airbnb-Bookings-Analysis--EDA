# **Q1.What can we learn about different hosts and areas?**

```python
# Choose Host_name and Neighbourhood_group

host_area = new_data.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
top_host = host_area.sort_values(by=['calculated_host_listings_count'], ascending = False).head(-5)
top_host
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(30,15))
box_plot=sns.boxplot(x='price',data=new_data).set_title("Price distribution by Area")
plt.xlabel('Price')
plt.ylabel('Area')

plt.show()
```

```python
# Let's find 10th percentile and 90 percentile
lower_quartile =new_data['price'].quantile(0.10)
lower_quartile
```

```python
upper_quartile =new_data['price'].quantile(0.90)
upper_quartile
```

```python
# new data rows = 49-269

new_data=new_data.drop(new_data[new_data['price']<lower_quartile].index)
new_data=new_data.drop(new_data[new_data['price']>upper_quartile].index)
new_data.head(-5)
```

```python
new_data.agg({'price':['mean','median']})
```
